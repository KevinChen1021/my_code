2. sorting
 input: array of n numbers
 output:sorted array

 2.1 permutation sort(stupid)
  list all the possible permutations that n numbers can generate(n factorial(n!))
  check if one permutation is sorted
 
 2.2 selection sort
  1.traverse the whole array
  2.find the largest
  3.swap it with the last one
  4.go 1-3 again 
  

```python
def prefix_max(Array, i):
    '''return index of maximum in Array[0:i+1]'''
    if i > 0:
        j = prefix_max(Array, i-1)
        if Array[j] > Array[i]:
            return j
    return i

def selection_sort(Array, i = None):
    '''sort Array[0: i+1]'''
    if i is None:
        i = len(Array) - 1
    if i > 0:
        j = prefix_max(Array, i)
        Array[i], Array[j] = Array[j], Array[i]
        selection_sort(Array, i-1)
```

  2.3 merge sort (O(nlogn))
   1.merge sort
    put same amount of numbers into a box
    merge two boxes
    doing merge sort

   2.merging
    two pointers, put the largest into the end of the new empty array
    "largest pointer" goes one unit backward

```python
def merge_sort(Array, a = 0, b = None):
    '''sort Array[a:b]'''
    if b is None:
        b = len(Array)
    if a < b - 1:
        c = (a + b + 1) // 2
        merge_sort(Array, a, c)
        merge_sort(Array, c, b)
        L, R = Array[a:c], Array[c:b]
        merge(L, R, Array, len(L), len(R), c, b)

def merge(L, R, Array, i, j, a, b):
    '''merge sorted L and R into Array[a:b]'''
    if a < b:
        if (j<=0) or (i > 0 and L[i-1] > R[j-1]):
            Array[b-1] = L[i-1]
            i -= 1
        else:
            Array[b-1] = R[j-1]
            j -= 1
        merge(L, R, Array, i, j, a, b-1)
```
